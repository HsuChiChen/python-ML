````python
# 匯入 pandas 模組
import pandas as pd

# 匯入 numpy 模組
import numpy as np

import re
def bprint(*args, t: str="=") -> None:
    start = "-"*30 + t + "-"*30
    end = "-"*(60+len(t))
    print(re.sub('[a-z]=[a-z]', ' = ', '\n'.join([start, *args, end])))
````

### Series

`pandas.Series` 為**維度為 1**且**資料型態皆相同**的結構。

- 操作方法與操作 `numpy.ndarray` 陣列相同
    - 使用 `[位置]` 取值
    - 使用 `[起始位置:結束位置]` 取連續值
        - 包含起始位置
        - **不包含**結束位置
- 可以給予每個 `pandas.Series` 位置中的值一個名字
    - 使用 `index` 參數進行命名
    - 使用 `[index 名稱]` 取值
    - 使用 `[起始 index 名稱:結束 index 名稱]` 取連續值
        - 包含起始位置
        - **包含**結束位置
- 可以給予 `pandas.Series` 一個名字
    - 使用 `name` 參數進行命名
- 透過給予名稱讓資料具有**可閱讀性**

````python
# 宣告 int64(default) Series 變數
s1 = pd.Series(data=[1, 2, 3])    

bprint(
    f"s1=\n{s1}\n",
    f"{s1.dtype=}\n",    # dtype('int64')
    f"{s1.shape=}",    # (3,)
    t="Series s1"
)


# 宣告 float64(default) Series 變數
s2 = pd.Series([1., 2., 3.])

bprint(
    f"s2=\n{s2}\n",
    f"{s2.dtype=}\n",  # dtype('int64')
    f"{s2.shape=}",    # (3,)
    t="Series s2"
)


# 宣告 Series 變數 (使用index, name)
s3 = pd.Series(
    name='series s3',
    data=[1, 2, 3,],
    index=['1st', '2nd', '3rd']
)

bprint(
    f"s3=\n{s3}\n",
    f"{s3.name=}\n",       # series s3
    f"{s3.index=}",        # ['1st', '2nd', '3rd']
    t="Series s3"
)

bprint(
    f"{s3[0]=}",                                "\t輸出 位置 0 的值",
    f"{s3['1st']=}",                            "\t輸出 名稱為 1st 的值",
    f"{s3[1]=}",                                "\t輸出 位置 1 的值",
    f"{s3['2nd']=}",                            "\t輸出 名稱為 2nd 的值",
    f"{s3[2]=}",                                "\t輸出 位置 2 的值",
    f"{s3['3rd']=}",                            "\t輸出 名稱為 3rd 的值\n",

    f"s3[:2]=\n{s3[:2]}",                       "\t輸出 從 '頭(0)' 到 '2的前一個(1)' 的所有值\n",
    f"s3['1st':'3rd']=\n{s3['1st':'3rd']}",     "\t輸出 從 '1st' 到 包含'3rd' 的所有值",

    t="s3 indexing"
)


# 宣告 Series 變數 (使用dict)
s4 = pd.Series(
    name="series s4",
    data={
        "A": 1,
        "B": 2,
        "C": 3
    }
)
bprint(
    f"s4=\n{s4}\n",
    f"{s4.name=}\n",       # series s4
    f"{s4.index=}",        # ['1st', '2nd', '3rd']
    t="Series s4"
)
````

### DataFrame

`pandas.DataFrame` 為**維度為 2**，**資料型態不一定相同**的結構。

- 由一個或多個 `pandas.Series` 所組成
- 以**表格**的方式呈現與操作（如同 SQL）
    - 使用**行（row）** 與**列（column）** 進行操作
    - 每列都會自動被轉換成 `pandas.Series`
- 取值方法
    - 先選擇**行（column）**，再使用**列的位置（row position）**取值
        - 同 `pandas.Series` 的操作方式
    - 先選擇**列（row position）**，再使用**行的位置（column position）** 或**名稱（column index）** 取值
        - 必須使用 `pandas.DataFrame.iloc[位置]`
    - 先選擇**列的名稱（row index）**，再使用**行的位置（column position）** 或**名稱（column index）** 取值
        - 使用 `index` 參數進行命名
        - 必須使用 `pandas.DataFrame.loc[名稱]`
    - 支援連續取值，與 `pandas.Series` 操作方法相同
- 透過給予名稱讓資料具有**可閱讀性**

````python
# DataFrame

# 宣告 DataFrame 變數
df1 = pd.DataFrame({ 
    'col1': [1, 2, 3, 4],
    'col2': [5., 6., 7., 8.],
    'col3': ['a', 'b', 'c', 'd']
})

bprint(
    f"df1=\n{df1}",                             "\t輸出 DataFrame df1\n",
    f"df1.dtypes=\n{df1.dtypes}",               "\t輸出 DataFrame df1 的 dtypes\n",
    f"{df1.shape=}",                            "\t輸出 DataFrame df1 的 shape\n",

    # could access by . (dot operator). (i.e. df1.col1 is df1['col1'])
    f"df1['col1']=\n{df1['col1']}",             "\t輸出 DataFrame df1 名稱為 col1 的行\n",
    f"{type(df1['col1'])=}",                    "\tdf1['col1'] 型態為 Series",
    t="DataFrame df1"
)

# 宣告 DataFrame 變數 (by data, columns claim)
df2 = pd.DataFrame(           
    columns=['col1', 'col2', 'col3', 'col4'],
    data=[[1, 2 ,3 ,4],
          [5., 6., 7., 8.],
          ['a', 'b', 'c', 'd']],
    index=['row1', 'row2', 'row3']
)

bprint(
    f"df2=\n{df2}",                             "\t輸出 DataFrame df2\n",
    f"df2.info=\n{df2.info}",                   "\t輸出 DataFrame df2 的 資訊\n",
    f"df2=\n{df2}",                             "\t輸出 DataFrame df2\n",
    t="DataFrame df2"
)
````

### 資料前處理（Data Preprocessing）

有時後部份資料會遺失或是有誤（例如型態錯誤），這時就必須要進行資料前處理。

在 `pandas` 中不存在或遺失的值以 `None` 表示，若資料欄位為浮點數則改用 `numpy.nan` 表示。

|函數|意義|
|-|-|
|`pandas.(DataFrame\|Series).isna`|列出不存在或遺失的資料|
|`pandas.(DataFrame\|Series).notna`|列出存在的資料|
|`pandas.(DataFrame\|Series).dropna`|去除不存在或遺失的資料|
|`pandas.(DataFrame\|Series).fillna`|對不存在或遺失的資料覆寫指定數值|

處理遺失資料的策略為

1. 找到遺失的資料
2. 若與該資料相同的數值有某些**簡單**或是**可預測**的值，則以**覆寫**遺失值
    - 若資料為**數字**，則可以填入**平均值**或是**中位數**
    - 若資料為**類別**，則可以填入**出現次數最多的類別**
3. 若無法處理，則**丟棄**資料

````python
# 資料前處理

# 宣告 DataFrame 變數
df4 = pd.DataFrame(
    data={
        'hw1': [90, 90, 90, np.nan, 70],
        'hw2': [90, 90, np.nan, np.nan, 70],
        'hw3': [90, 90, np.nan, np.nan, 70],
        'final': [90, 90, np.nan, np.nan, 70],
  	},
	index=[
        'Alice', 'Bob', 'Cindy', 'Emmily', 'Frank'
    ],
)

bprint(
    f"df4.notna()=\n{df4.notna()}",     "\t找出 DataFrame df4 中沒有遺失資料的欄位\n",
    f"df4.isna()=\n{df4.isna()}",       "\t找出 DataFrame df4 中遺失資料的欄位\n",
    
    f"df4['hw1'].fillna(0)=\n"
    f"{df4['hw1'].fillna(0)}",          "\t對 df4['hw1'] 中遺失的欄位填入0\n",

    f"df4['final'].dropna()=\n"
    f"{df4['final'].dropna()}",         "\t移除 df4['final'] 中遺失的欄位\n",
    t="DataFrame df4"
)
````

### 取得資料（Select Data）

使用條件運算子對 `pandas.Series` 進行運算，並回傳 `dtype` 為 `bool` 的 `pandas.Series`。

可以進一步在使用回傳的 `pandas.Series` 取出 `pandas.DataFrame` 中的符合條件的資料列（row）。

````python
# 取得資料

# 宣告 DataFrame 變數
df5 = pd.DataFrame({              
    # 名字
    'name': ['Alice', 'Beth',
             'Carl', 'Dennis'],   
    # 年紀
    'age': [18, 30, 24, 40],      
    # 性別
    'sex': ['F', 'F', 'M', 'M'],  
})

bprint(
    f"df5['age'] > 25\n{df5.age > 25}",                    "\t找出 大於25歲 的資料\n",
    f"df5['sex'] == 'F'\n{df5.sex == 'F'}",                "\t找出 為女性 的資料\n",

    f"df5[df5['age'] > 25]\n{df5[df5.age > 25]}",          "\t取出 大於25歲 的資料\n",
    f"df5[df5['sex'] == 'F']\n{df5[df5.sex == 'F']}",      "\t取出 為女性 的資料\n",

    "(df5['age'] > 25)\n&\n(df5['sex'] == 'F')\n"
    f"{(df5['age'] > 25) & (df5['sex'] == 'F')}",           "\t找出年紀大於 25 歲且性別為女性的資料\n",

    "df5[(df5['age'] > 25)\n\t&\n\t(df5['sex'] == 'F')]\n"
    f"{df5[(df5['age'] > 25) & (df5['sex'] == 'F')]}",      "\t取出年紀大於 25 歲且性別為女性的資料\n",
)
````


### 資料運算（Data Operation）

使用 `pandas` 內建的運算功能進行資料運算，基本上與 `numpy.ndarray` 所提供的功能相同。

若需要客制化每一筆資料的運算，則使用 `pandas.(DataFrame|Series).apply`。

- `axis=0` 為行（column）方向運算，預設為 `axis=0`
- `axis=1` 為列（row）方向運算

|函數|意義|
|-|-|
|`pandas.(DataFrame\|Series).add`|每筆資料進行加法|
|`pandas.(DataFrame\|Series).sub`|每筆資料進行減法|
|`pandas.(DataFrame\|Series).mul`|每筆資料進行乘法|
|`pandas.(DataFrame\|Series).div`|每筆資料進行除法|
|`pandas.(DataFrame\|Series).max`|取出資料中最大值|
|`pandas.(DataFrame\|Series).min`|取出資料中最小值|
|`pandas.Series.argmax`|取出資料中最大值的位置|
|`pandas.Series.argmin`|取出資料中最小值的位置|
|`pandas.Series.idxmax`|取出資料中最大值的名稱|
|`pandas.Series.idxmin`|取出資料中最小值的名稱|
|`pandas.(DataFrame\|Series).mean`|計算平均數|
|`pandas.(DataFrame\|Series).var`|計算變異數|
|`pandas.(DataFrame\|Series).std`|計算標準差|
|`pandas.(DataFrame\|Series).sum`|計算總和|

````python
# 資料運算

# 宣告 DataFrame 變數
df6 = pd.DataFrame( 
    {
        'year': [1995, 1989, 1947],
        'month': [10, 6, 2],
        'date': [12, 4, 28],
    }, 
    index=['my birthday', 'a big day', 'another big day']
)

# 將 DataFrame df6 中所有數值加 3
print(df6.add(3))            
print()
# 將 DataFrame df6 中所有數值減 5
print(df6.sub(5))            
print()
# 將 DataFrame df6 中所有數值乘 7
print(df6.mul(7))            
print()
# 將 DataFrame df6 中所有數值除 9
print(df6.div(9))            
print()
# 找出 DataFrame df6 中每一行的最大值
print(df6.max())             
print()
# 找出 DataFrame df6 中每一行的最小值
print(df6.min())             
print()
# 找出 Series df6['year'] 中最大值的位置
print(df6['year'].argmax())  
# 找出 Series df6['month'] 中最小值的位置
print(df6['month'].argmin()) 
# 找出 Series df6['year'] 中最大值的名稱
print(df6['year'].idxmax())  
# 找出 Series df6['date'] 中最小值的名稱
print(df6['date'].idxmin())  
print()
# 計算 DataFrame df6 中每一行的平均值
print(df6.mean())            
print()
# 計算 DataFrame df6 中每一行的變異數
print(df6.var())             
print()
# 計算 DataFrame df6 中每一行的標準差
print(df6.std())             
print()
# 計算 DataFrame df6 中每一行的總和
print(df6.sum())             
````

````python
import datetime

# 宣告 DataFrame 變數
df7 = pd.DataFrame({                
    # 名字
    'name': ['Alice', 'Beth',
             'Carl', 'Dennis'],     
    # 年紀
    'age': [18, 30, 24, 40],        
    # 性別
    'sex': ['F', 'F', 'M', 'M'],    
})

# 計算出生西元年
def birth_year(age: int) -> int:
    return datetime.date.today().year - age

# 根據性別輸出 Female 或 Male
def sex_str(sex: str) -> str:                   
    return 'female' if sex == 'F' else 'male'

# 將資料轉換成可閱讀的資訊
def info_str(row: pd.Series) -> str:
    return '{} is a {} and was born at {}.'.format(
        row['name'],
        sex_str(row['sex']),
        birth_year(row['age'])
    )

# 輸出出生西元年
print(df7['age'].apply(birth_year)) 
print()
# 輸出性別
print(df7['sex'].apply(sex_str))    
print()
# 輸出可閱讀的資訊
print(df7.apply(info_str, axis=1))  
````

## 其他常用操作

1. 合併操作

|函數|意義|
|-|-|
|`pandas.concat(Iterable[DataFrame])`|合併DataFrame|
|`pandas.(DataFrame\|Series).reset_index()`|重置index|

````python
columns = list('ABC')

df11 = pd.DataFrame(
    data=np.arange(6).reshape(2, 3),
    columns=columns
)
df12 = pd.DataFrame(
    data=np.arange(6).reshape(2, 3)+10,
    columns=columns
)

bprint(
    f"pd.concat([df11, df12], axis=1)\n{pd.concat([df11, df12], axis=1)}", "\t往右長 axis=1\n",
    f"pd.concat([df11, df12], axis=0)\n{pd.concat([df11, df12], axis=0)}", "\t往下長 axis=0",
    t="concat"
)
df13 = pd.concat([df11, df12], axis=0)

bprint(
    f"df13=\n{df13}\n",
    f"df13.reset_index()\n{df13.reset_index()}",             "\t重置 index值\n",
    f"df13.reset_index(drop=True)\n{df13.reset_index(drop=True)}",    "\t重置 index值 並 刪除之前的index",
    t="reset_index"
)
````

2. 逐列操作

|函數|意義|
|-|-|
|`pandas.(DataFrame\|Series).map(Callable)`|對每列進行映射(map)操作|
|`pandas.(DataFrame\|Series).apply(Callable)`|對每列進行函數(function)操作|

````python
from datetime import datetime, timedelta

num_data = 10
clients = ["student",  "teacher", "others"]
gender = ["male",  "female"]
color = ["white", "black", "yellow"]

factory = lambda labels: [labels[i] for i in np.random.randint(0, len(labels), num_data)]

df14 = pd.DataFrame({
    "height": factory(range(150, 190)), 
    "weight": factory(range(40, 90)), 
    "age": factory(range(18, 75)), 
    "clientType": factory(clients), 
    "gender": factory(gender), 
    "tshirtColor": factory(color),
    "registerTime": factory(range(-5, 5))
})

# 建立clients編碼.
client_map = dict([(v, i) for i, v in enumerate(clients)])

# relative date to absolute date.
r2a = lambda dt: datetime.today().date() + timedelta(int(dt))

bprint(
    f"df14\n{df14}\n",

    f"{client_map=}\n",
    "df14.clientType.map(client_map)\n"
    f"{df14.clientType.map(client_map)}", "\t將 clientType 編碼\n",

    "df14.registerTime.apply(r2a)\n"
    f"{str(df14.registerTime.apply(r2a))}", "\t將relative date 轉為 absolute date\n",
)

# pd.to_datetime(df14.registerTime)
````
